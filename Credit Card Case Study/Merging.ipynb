```python
import pandas as pd

students = pd.DataFrame(
    {
        "Student_id" : [101,102,103,104],
        "Name" : ["Joy","Rohan","Toy","Roy"],
        "Age" : [20,21,19,22]
    }
)

Scores = pd.DataFrame({
    "Student_id" : [101,102,103,105],
    "Subject" : ["Maths", "Science","Hindi","English"],
    "Scores" : [90,98,89,78]

})

print(students)
print(Scores)
```

```python
# Inner Join

inner_join = pd.merge(students,Scores,on='Student_id')

inner_join
```

```python
# Left Join

Left_join = pd.merge(students,Scores,on='Student_id', how="left")

Left_join
```

```python
#Right Join

Right_join = pd.merge(students,Scores,on='Student_id', how="right")

Right_join
```

```python
# Outer Join

Outer_join = pd.merge(students,Scores,on='Student_id', how="outer")

Outer_join
```
